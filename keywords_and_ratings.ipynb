### データの抜き出し、整形

### keywords_vote... id,keywords,vote_average

```python
keywords_vote=pd.merge(keywords,vote_average,on="id")
keywords_vote.head()
```

### vote=0,NaN / keywords=[ ]のデータを除外

```python
#NaN削除
keywords_vote.dropna()
```

```python
#[]削除
keywords_vote=keywords_vote.drop(keywords_vote[keywords_vote.keywords=="[]"].index.values,axis=0)
keywords_vote
```

```python
#ratings 0 削除
keywords_vote=keywords_vote.drop(keywords_vote[keywords_vote.vote_average==0].index.values,axis=0)
keywords_vote
```

### keywords_listづくり

```python
#その名の通りstr->list

def make_list_from_str(string):
    if isinstance(string, str):
        result_list = eval(string)
    else:
        result_list = np.nan
    return result_list
```

```python
#keywords_vote["keywords_list"]がlist化されたkeywords
keywords_vote["keywords_list"] = keywords_vote.keywords.map(make_list_from_str)
keywords_vote = keywords_vote.dropna()
keywords_vote.info()
```

### ダミー変数作成

```python
##ジャンルのdictから、それぞれのidだけを抜き出したリストの作成
def extract_ids(genres_list):
    id_list = []
    for genre in genres_list:
        id_list.append(genre["id"])
    return id_list
```

```python
##ジャンルのdictから、それぞれのnameだけを抜き出したリストの作成
def extract_names(genres_list):
    id_list = []
    for genre in genres_list:
        id_list.append(genre["name"])
    return id_list
```

```python
keywords_vote["keywords_name"] = keywords_vote["keywords_list"].map(extract_names)
keywords_vote.head()
```

```python
keywords_vote["keywords_id"] = keywords_vote["keywords_list"].map(extract_ids)
keywords_vote.head()
```

```python
keywords_vote["keywords_count"] = keywords_vote["keywords_list"].map(lambda x: len(x))
keywords_vote.head()
```

```python
#空のデータフレーム
keywords_only=pd.DataFrame(index=['index'],columns=['name','num'])
#追加する行をSeriesでつくり、追加していく
for i in keywords_vote.index:
    for j in range(len(keywords_vote.keywords_list[i])):
        if keywords_vote.keywords_list[i][j]["id"] in keywords_only.index:
            keywords_only.num[keywords_vote.keywords_list[i][j]["id"]]+=1
        else:
            df_buffer=pd.DataFrame({'name':keywords_vote.keywords_list[i][j]["name"],'num':1.},index=[keywords_vote.keywords_list[i][j]["id"]])
            keywords_only = keywords_only.append(df_buffer)
            #print(keywords_only.count[0])
            
keywords_only
```

```python
keywords_only=keywords_only.dropna()
```

```python
#sklearn インポートのヒント　https://github.com/scikit-learn/scikit-learn/issues/6161
from sklearn.model_selection import train_test_split

# モデル
from sklearn import linear_model

# モデルのインスタンス
l_model = linear_model.LinearRegression()
 
# 説明変数に "price" 以外を利用
X = dummies_vote.drop(["vote_average","id"], axis=1)

# 目的変数
Y = dummies_vote["vote_average"]

# 学習データとテストデータ分ける(test_sizeはテストデータの割合)
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.3,random_state=0)

# モデルのあてはめ
clf = l_model.fit(X_train,y_train)
print("train:",clf.__class__.__name__ ,clf.score(X_train,y_train))
print("test:",clf.__class__.__name__ , clf.score(X_test,y_test))
 
# 偏回帰係数
print(pd.DataFrame({"Name":X.columns,
                    "Coefficients":clf.coef_}).sort_values(by='Coefficients') )
 
# 切片 
print(clf.intercept_)
```
